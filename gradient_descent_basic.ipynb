```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
m = 78.35063668
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# let m = 78.35063668
# let assume for intercept  b = 0
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,m*X+0,color='green',label='SGD')
plt.legend()
plt.show()
```

```python
m = 78.35063668
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() -b)
loss_slope
```

```python
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,m*X+0,color='green',label='SGD')
plt.plot(X,m*X+b,color='yellow',label='SGD_S1')
plt.legend()
plt.show()
```

```python
# Iteration 2

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,m*X+0,color='green',label='SGD')
plt.plot(X,m*X+b,color='yellow',label='SGD_S2')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    
    b = b - (lr * loss_slope)

    y_pred = m*X + b

    plt.plot(X,y_pred)
plt.scatter(X,y)
```
