```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.sample(5)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

## Bar plot

```python
sns.barplot(data=titanic,x='Pclass',y='Age',hue='Sex')
```

## Box Plot

```python
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

## Distplot

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

## HeatMap (cat - cat)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']),annot=True)
```

```python
titanic.groupby('Sex').mean()['Survived']*100
```

## Clustermap ( Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']), cmap='YlGnBu')
```

## Pair Plot

```python
sns.pairplot(iris,hue='species',palette='YlGnBu')
```

## LinePlot (Num - NUm)

```python
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(data=new,x='year',y='passengers')
```
