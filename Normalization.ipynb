```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
```

```python
df.columns = ['Class label','Alcohol','Malic acid']
```

```python
df
```

```python
sns.kdeplot(data=df , x='Alcohol')
```

```python
sns.kdeplot(data=df,x='Malic acid')
```

```python
color_dict = {1:'red',3:'green',2:'blue'}
sns.scatterplot(df,x='Alcohol',y='Malic acid',hue='Class label',palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label',axis=1), df['Class label'], test_size=0.2, random_state=0)

X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns = X_test.columns)
```
